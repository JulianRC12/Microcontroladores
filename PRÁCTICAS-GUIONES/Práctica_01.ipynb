### <font color='blue'>**Convertir a escalas de grises**

```python
# Convertir a escalas de grises de forma correcta
# 1 Investigar las librerías y modulos necesarios para la conversión, para nuestro caso:
from skimage import color

# 2 Cargar una imagen de prueba
coffe_image = data.coffee()

# 3 Convertir la imagen a escala de grises usando el metodo correcto de la librería
gray_scaled_coffe_image = color.rgb2gray(coffe_image)
```

```python
# 4  Visualizar la imagen original utilizando pyplot
plt.imshow(coffe_image)
plt.title('taza de café')
plt.axis('off')
```

```python
# 5  Visualizar la imagen convertida a escala de grises utilizando pyplot
plt.imshow(gray_scaled_coffe_image, cmap='gray')
plt.title('taza de café',)
plt.axis('off')
```

### <font color='blue'>**Separación de canales RGB**

```python
# Uso de numpy para separar los canales de la imagen RGB
# 1 Investigar las librerías y modulos necesarios para la conversión

# 2 Cargar una imagen de prueba
cat_image = data.cat()

# 3 Verificar las dimensiones de la imagen
print(cat_image.shape)
```

```python
# 4 Obtener los valores del canal rojo de la imagen
red_channel = cat_image[:,:,0]

# 5 Obtener los valores del canal verde de la imagen
green_channel = cat_image[:,:,1]

# 6 Obtener los valores del canal azul de la imagen
blue_channel = cat_image[:,:,2]
```

```python
#6 Visualizar en una solo gráfica los tres canales usando matplotlib
plt.figure(figsize=(18,6))
plt.subplot(1,3,1); plt.imshow(red_channel, cmap=plt.cm.Reds);plt.axis('off')
plt.subplot(1,3,2); plt.imshow(green_channel, cmap=plt.cm.Greens);plt.axis('off')
plt.subplot(1,3,3); plt.imshow(blue_channel, cmap=plt.cm.Blues);plt.axis('off')
```

```python
# 7  Visualizar la imagen original utilizando pyplot
plt.imshow(cat_image)
plt.title('gato',)
plt.axis('off')
```
